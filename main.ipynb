```python
def clean_text(txt):
    txt = txt.lower()
    txt = re.sub(r"i'm", "i am", txt)
    txt = re.sub(r"he's", "he is", txt)
    txt = re.sub(r"she's", "she is", txt)
    txt = re.sub(r"that's", "that is", txt)
    txt = re.sub(r"what's", "what is", txt)
    txt = re.sub(r"where's", "where is", txt)
    txt = re.sub(r"\'ll", " will", txt)
    txt = re.sub(r"\'ve", " have", txt)
    txt = re.sub(r"\'re", " are", txt)
    txt = re.sub(r"\'d", " would", txt)
    txt = re.sub(r"won't", "will not", txt)
    txt = re.sub(r"can't", "can not", txt)
    txt = re.sub(r"[^\w\s]", "", txt)
    return txt
```

```python
question= []
answer= []

for line in input:
  question.append(clean_text(line))

for line in output:
  answer.append(clean_text(line))
```

```python
word2count= {}

for line in question:
  for word in line.split():
    if word not in word2count:
      word2count[word]= 1
    else:
      word2count[word]+= 1

for line in answer:
  for word in line.split():
    if word not in word2count:
      word2count[word]= 1
    else:
      word2count[word]+= 1

word2count
```

```python
thresh =3
vocab= {}
word_num= 0

for word, count in word2count.items():
  if count >= thresh:
    vocab[word]= word_num
    word_num+=1

vocab
```

```python
# for i in range(len(question)):
#   question[i]= '<SOS> '+question[i]+' <EOS>'

for i in range(len(answer)):
  answer[i]= '<SOS> '+answer[i]+' <EOS>'
```

```python
tokens= ['<PAD>', '<EOS>', '<OUT>', '<SOS>']
x= len(vocab)

for token in tokens:
  vocab[token]= x
  x+= 1
```

```python
vocab['what']= vocab['<PAD>']
vocab['<PAD>']= 0
```

```python
encoder_inp= []

for line in question:
  lst= []
  for word in line.split():
    if word not in vocab:
      lst.append(vocab['<OUT>'])
    else:
      lst.append(vocab[word])
  encoder_inp.append(lst)

encoder_inp
```

```python
decoder_inp= []

for line in answer:
  lst= []
  for word in line.split():
    if word not in vocab:
      lst.append(vocab['<OUT>'])
    else:
      lst.append(vocab[word])
  decoder_inp.append(lst)

decoder_inp
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Embedding, LSTM, Input
```

```python
enc_inp = Input(shape=(15, ))
dec_inp = Input(shape=(15, ))
```

```python
VOCAB_SIZE = len(vocab)

embed = Embedding(VOCAB_SIZE+1, output_dim=20, input_length=15, trainable=True)
```

```python
enc_embed = embed(enc_inp)
enc_lstm = LSTM(400, return_sequences=True, return_state=True)
enc_op, h, c = enc_lstm(enc_embed)
enc_states = [h, c]
```

```python
dec_embed = embed(dec_inp)
dec_lstm = LSTM(400, return_sequences=True, return_state=True)
dec_op, _, _ = dec_lstm(dec_embed, initial_state=enc_states)
```

```python
dense = Dense(VOCAB_SIZE, activation='softmax')
```

```python
dense_op = dense(dec_op)
```

```python
model = Model([enc_inp, dec_inp], dense_op)
```

```python
model.compile(loss='categorical_crossentropy',metrics=['acc'],optimizer='adam')
```

```python

model.fit([encoder_inp, decoder_inp],decoder_final_output,epochs=40)
```
